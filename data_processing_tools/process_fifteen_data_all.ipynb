```python
import re
def extract_answer_from_clue(clue):
    regex = '\d*\.*\s{1,2}([A-Z\s\-]{2,}\s)'
    try:
        answer = re.search(regex,clue).group(0)
        return clue.replace(answer,'')
    except Exception:
        return clue
    
```

```python
def remove_prefix_number_from_clue(clue):
    regex = '(\d{1,2}\.*)\s'
    try:
        answer = re.search(regex,clue).group(0)
        return clue.replace(answer,'')
    except Exception:
        return clue
    
```

```python
df['clue'] = df['clue'].apply(extract_answer_from_clue)
```

```python
df['clue'] = df['clue'].apply(remove_prefix_number_from_clue)
```
