```python
import pandas as pd
import re
```

```python
df = pd.read_csv("../fifteen_dataset/fifteen_all_tagged_3.csv")
```

```python
df['defs'] = df['clue'].apply(lambda x:re.findall("(\|[\w\s]+\|)",x))
```

```python
df = df[df['defs'].apply(lambda x:len(x)>0)]
```

```python
df['sol'] = df['clue'].apply(lambda x:re.findall("([A-Z\s]{3,}\s)",x))
```

```python
df['ans_len'] = df['clue'].apply(lambda x:re.findall("\(\d[,-]*\d*\)",x))
```

```python
df['clue'] = df['clue'].apply(lambda x:re.sub("\(\d[,-]*\d*\)","",x))
```

```python
df['clue'] = df['clue'].apply(lambda x:re.sub("([A-Z\s]{3,}\s)","",x))
```

```python
df['clue'] = df['clue'].apply(lambda x:re.sub("^\d{1,2}","",x))
```

```python
df['clue'] = df['clue'].apply(lambda x:re.sub("\|","",x))
```

```python
df['defs'] = df['defs'].apply(lambda x:[w.replace('|','') for w in x])
```

```python
df['defs'] = df['defs'].apply(lambda x:[w.split(' ') for w in x])
```

```python
df['defs'] = df['defs'].apply(lambda x:[item for sublist in x for item in sublist])
```

```python
df.reset_index(drop=True)
```

```python
def clue2SeqTags(clue,defs):
    arr = [word for word in clue.split(' ') if word != ""]
    seq_tags = []
    for word in arr:
        if word not in defs:
            seq_tags.append(0)
        else:
            seq_tags.append(1)
    return seq_tags
    
```

```python
seq_tags = []
for i in range(len(df)):
    seq_tags.append(clue2SeqTags(df.iloc[i].clue,df.iloc[i].defs))
```

```python
df['seq_tags'] = seq_tags
```
